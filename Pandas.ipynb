#Data Cleaning

```python
df0 = pd.read_csv('CarPrice.csv')
df0.head()
```

```python
# remove empty cells
df = df0.dropna()

#replace null with something (maybe 0 or mean/median)
carlength = df['carlength'].mean()
df.fillna({'carlength':carlength}, inplace=True)
```

# Trick Concepts

```python
# groupby

# split data into buckets -> do something each bucket -> combine
df.groupby("carbody")["price"].mean()
```

```python
# agg()

df.groupby("carbody")["price"].agg('mean')
# .mean() same as .agg('mean')
```

```python
# transform
df['carbody_avg'] = df.groupby('carbody')['price'].transform('mean')
df['carbody_avg'][:5]

# difference is, it broadcasts group statistic back to rows, so returns same length as original df
# good for filtering rows or create feaures or compare group vs rows
```

# Trick Questions

## Group-wise normalization

```python
df['price_norm_by_carbody'] = (
    df.groupby('carbody')['price']
    .transform(lambda x: (x-x.mean())/x.std())
)
# bucket by bodytype,
# operate on price (in each bucket),
# transform price such that ....
```

```python
df['price_norm_by_carbody'][:5]
```

## Conditional aggregation inside groupby

```python
df.groupby('fueltype').apply(
    lambda g: g.loc[g['price'] > g['price'].median(), "horsepower"].mean(),
    include_groups=False
)

# avoid using apply, instead use transform + filter + groupby
```

```python
# improved version (avoids apply which is slower ... coz less vectorized)

median_price = df.groupby("fueltype")["price"].transform("median")

df.loc[df["price"] > median_price] \
  .groupby("fueltype")["horsepower"] \
  .mean()
```

## Multi-column ranking within groups

```python
df["hp_rank"] = df.groupby("fueltype")["horsepower"].rank(ascending=False)
```

```python
df["combo_rank"] = (
    df.groupby("fueltype")[["horsepower", "price"]]
      .apply(lambda g: g.rank(pct=True).mean(axis=1))
      .reset_index(level=0, drop=True)
)

df["combo_rank"][:5]
```

##Pivot Table

```python
pd.pivot_table(
    df,
    values="price",
    index="carbody",
    columns="drivewheel",
    aggfunc=lambda x: x.quantile(0.5),
)
```

## Window operations within Groups

```python
df["rolling_hp"] = (
    df.sort_values("price")
      .groupby("fueltype")["horsepower"]
      .transform(lambda x: x.rolling(3, min_periods=1).mean())
)

df["rolling_hp"][:5]
```

##Joins

```python
stats = df.groupby("carbody")["price"].agg(["mean", "std"])
stats
```

```python
df = df.join(stats, on="carbody")
df["zscore"] = (df["price"] - df["mean"]) / df["std"]
df["zscore"][:5]
```

##Boolean Masking

```python
# method I
df.groupby("carbody")["price"].apply(lambda x: (x > x.mean()).sum())
```

```python
# method II
(df["price"] > df.groupby("carbody")["price"].transform("mean")).groupby(df["carbody"]).sum()
```
